```python
import requests
from bs4 import BeautifulSoup

res = requests.get('http://info.cern.ch')
soup = BeautifulSoup(res.content, 'html.parser')

# text
print(soup.a.get_text())
# attribute - dictionary로 생각하면 됨.
print(soup.a['href'])
```

```python
# 태그를 지우고 전체 텍스트를 가져옴.
print(soup.get_text())
```

```python
print(type(soup.a))
soup.a
```

```python
links = [tag['href'] for tag in soup.select('a')]
links
```

```python
texts = [tag.get_text() for tag in soup.select('a')]
texts
```

```python
def mean(*args): 
    return sum(args) / len(args)

print(mean(1, 2, 3, 4))
m = [1, 2, 3, 4, 5, 6]
print(mean(*m))
```

```python
d1 = dict(a=1, b=2, c=3)
d1 = {'a': 1, 'b': 2, 'c': 3}
d1['d'] = 5

print(d1)
d2 = dict(d=3, e=4, **d1)
print(d2)
```

```python
students = [
    {'name': 'sookbun', 'points': [100, 20]},
    {'name': 'duru', 'points': [120, 200]},
    {'name': 'egoing', 'points': [100, 35]},
]

students = [dict(sum=sum(s['points']), **s) for s in students]

import pprint
pprint.pprint(students)
```
